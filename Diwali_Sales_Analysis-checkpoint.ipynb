```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# drop unrelated/ blank column
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)  # here inplace=True is asingning data  after null value removal to df and we got all not null values.
```

```python
# Change datatype to int 

df['Amount']=df['Amount'].astype('int')
```

```python
# rename columns
df.rename(columns={'Marital_Status':'Sadisuda'})   #it is note reflected after this cell because we are note assinging it to dataframe
```

```python
# Exploratory data Analysis(EDA)------>

# Gender------>
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
# if we wants value in bars -->
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# wants spending amount based on gender 
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#ploting this dat of genders

sns.barplot(x='Gender',y='Amount',data=sales_gender)

# Summary--> Most buyers are females and even purchasing power of female is greater than the mans.
```

```python
df.columns
```

```python
# Age----------------------->
# hue gives the separation based on given value (here for gender---> male and female)
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
# Total amount vs Age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
#Summary--->Most of buyers are the age group of 26-35  yrs female
```

```python
df.columns
```

```python
# state ------------->
# Total number of orders from Top 10 states.
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# Amount  from Top 10 states.
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)

# Summary--> Here we see the most of the orders and Amount generated from UP,Maharashtra and Karnataka respectivly.
```

```python
#Marital Status----->
sns.set(rc={'figure.figsize':(7,5)})
ax=sns.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_marital,hue='Gender')

# Summary-----> We can see the most of the buyers are the married womans and their purchasing power is also high.
```

```python
# Occupation-------------->

sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_marital,hue='Gender')

#Summary--->most of the buyers womans are working at IT sector,Healthcare and Aviation sector. 
```

```python
df.columns
```

```python
# Product Category------------------------------------>

sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_marital)

# Summary---> sales of clothing & Apparel are high but High amount is spended on food by buyers.
# Most sold products are Food,Clothing,Electronic gadgets and footwears.
```

```python
# Top Salling Products----->
sales_marital=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_ID',y='Amount',data=sales_marital)
```

```python
# Conclusion----->
# Married woman age group 26-35 from UP,Maharashtra,Karnataka working in IT ,HealthCare and Aviation are more likly to buy products from Food ,Clothing and Electronic Category.
```
